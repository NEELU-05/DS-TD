```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("DATA/covid19_Confirmed_dataset.csv")
dataset.head()
```

```python
dataset.shape
#df = dataset.drop(["Lat","Long"],axis=1,inplace=True)
dataset.head()
```

```python
CDA = dataset.groupby("Country/Region").sum()
CDA.head()
```

```python
CDA.shape
```

```python
CDA.loc["India"].plot()
CDA.loc["Spain"].plot()
CDA.loc["Germany"].plot()
CDA.
plt.legend()
plt.show()
```

```python
CDA.loc["India"][:30].plot()
plt.show()
CDA.loc["India"].diff().plot()
plt.show()
```

```python
display(CDA.loc["Spain"].diff().max())
display(CDA.loc["India"].diff().max())
display(CDA.loc["Germany"].diff().max())
```

```python
numeric_CDA = CDA.apply(pd.to_numeric, errors='coerce')

countries = list(numeric_CDA.index)
max_infected = []
for country in countries:
    max_infected.append(numeric_CDA.loc[country].diff().max())
CDA["max_infected"] = max_infected
```

```python
CDA
```

```python
corona_dataframe =pd.DataFrame(CDA["max_infected"])
display(corona_dataframe)
```

```python
happiness_report = pd.read_csv("DATA/worldwide_happiness_report.csv")
happiness_report
```

```python
usless_cols = ["Overall rank","Score","Generosity","Perceptions of corruption"]
happiness_report.drop(usless_cols,axis=1,inplace=True)
happiness_report
```

```python
happiness_report.set_index("Country or region",inplace=True)
happiness_report
```

```python
display(corona_dataframe.shape)
display(happiness_report.shape)
```

```python
data = corona_dataframe.join(happiness_report,how="inner")
data
```

```python
data.corr()
```

```python
data
```
