```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.plot(x,y,label ="pyplot",marker ="o")
plt.xlabel("x list"),plt.ylabel("y list")
plt.grid()
plt.show()
```

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
plt.bar(students,marks,color ="yellow")
plt.title("STUDENT MARKS")
plt.show()
```

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
plt.pie(revenue,labels= regions,explode=[0.1,0,0,0],autopct="%.2f")
plt.show()
```

```python
import random as r
sample_size = 1000
values = [r.randrange(1, 100) for _ in range(sample_size)] 
plt.hist(values,bins=20,color="skyblue")
plt.show()
```
